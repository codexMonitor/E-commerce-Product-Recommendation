```python
import pandas as pd

part1 = pd.read_csv("../data/raw/item_properties_part1.csv")
part2 = pd.read_csv("../data/raw/item_properties_part2.csv")
```

```python
item_properties = pd.concat(
    [part1, part2],
    ignore_index=True
)

print(item_properties.shape)
```

```python
categories = item_properties[
    item_properties["property"] == "categoryid"
].copy()

categories.head()
```

```python
categories = (
    categories
    .sort_values("timestamp")
    .drop_duplicates(
        subset="itemid",
        keep="last"
    )
)
```

```python
categories = categories.rename(
    columns={
        "value": "categoryid"
    }
)
```

```python
products = categories[
    ["itemid", "categoryid"]
].copy()
```

```python
products.head()
```

```python
products = products.sort_values("itemid")
```

```python
products.to_csv(
    "../data/processed/products.csv",
    index=False
)
```

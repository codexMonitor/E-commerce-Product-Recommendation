## Reused function

```python
def inspect_dataset(df, name):
    print("=" * 70)
    print(f"{name.upper()} DATASET")
    print("=" * 70)

    # Shape
    print("\n1. Shape")
    print(df.shape)

    # Columns
    print("\n2. Columns")
    print(df.columns.tolist())

    # Data types
    print("\n3. Data Types")
    display(df.dtypes)

    # Missing values
    print("\n4. Missing Values")
    missing = pd.DataFrame({
        "Missing": df.isnull().sum(),
        "Percentage": (df.isnull().mean()*100).round(2)
    })
    display(missing)

    # Duplicate rows
    print("\n5. Duplicate Records")
    print(df.duplicated().sum())

    # Summary
    print("\n6. Dataset Summary")
    display(df.describe(include="all"))

    # First rows
    print("\n7. First 5 Rows")
    display(df.head())

    print("\n")
```

## Inspect data

```python
inspect_dataset(events, "Events")

inspect_dataset(item_properties1, "Item Properties Part 1")

inspect_dataset(item_properties2, "Item Properties Part 2")

inspect_dataset(category_tree, "Category Tree")
```

# 2. Preprocessing

## Remove duplicate rows

```python
events = events.drop_duplicates()

print("Remaining rows:", len(events))
```

## Inspect NULL

```python
print(events.isnull().sum())
```

```python
invalid_transactions = events[
    (events["event"] == "transaction") &
    (events["transactionid"].isna())
]

print("Invalid transaction rows:", len(invalid_transactions))
```

## Check invalid timestamps

```python
events["datetime"] = pd.to_datetime(
    events["timestamp"],
    unit="ms",
    errors="coerce"
)
```

```python
invalid_time = events[
    events["datetime"].isna()
]

print(len(invalid_time))
```

```python
print(events["datetime"].min())
print(events["datetime"].max())
```

## Check invalid IDs

```python
invalid_visitors = events[
    events["visitorid"] <= 0
]

print(len(invalid_visitors))
```

```python
invalid_items = events[
    events["itemid"] <= 0
]

print(len(invalid_items))
```

```python
invalid_transaction_ids = events[
    (events["event"] == "transaction") &
    (events["transactionid"] <= 0)
]

print(len(invalid_transaction_ids))
```

```python
events = events.drop(invalid_visitors.index)

events = events.drop(invalid_transaction_ids.index)
```

## Drop helper column

```python
events = events.drop(columns=["datetime"])

events.head()
```

## Save cleaned dataset

```python
events.to_csv(
    "../data/processed/events_clean.csv",
    index=False
)

print("Saved successfully!")
```

# 4. Feature Engineering

```python
events = pd.read_csv("../data/processed/events_clean.csv")
```

Assign interaction weights

```python
weight_map = {
    "view": 1,
    "addtocart": 3,
    "transaction": 5
}

events["weight"] = events["event"].map(weight_map)

events.head()
```

Encode users and items

```python
from sklearn.preprocessing import LabelEncoder

user_encoder = LabelEncoder()
item_encoder = LabelEncoder()

events["user"] = user_encoder.fit_transform(events["visitorid"])
events["item"] = item_encoder.fit_transform(events["itemid"])
```

Save encoders

```python
import pickle

with open("../data/processed/user_encoder.pkl", "wb") as f:
    pickle.dump(user_encoder, f)

with open("../data/processed/item_encoder.pkl", "wb") as f:
    pickle.dump(item_encoder, f)
```

Aggregate repeated interactions

```python
interactions = (
    events
    .groupby(["user", "item"], as_index=False)
    ["weight"]
    .sum()
)

interactions.head()
```

Create interaction matrix

```python
from scipy.sparse import csr_matrix

interaction_matrix = csr_matrix(
    (
        interactions["weight"],
        (
            interactions["user"],
            interactions["item"]
        )
    )
)
```

```python
interaction_matrix
```

Save interaction matrix

```python
from scipy.sparse import save_npz

save_npz(
    "../data/processed/interaction_matrix.npz",
    interaction_matrix
)
```

Create Train/Test datasets

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(
    interactions,
    test_size=0.2,
    random_state=42
)
```

```python
train, test
```

```python
train.to_csv(
    "../data/processed/train.csv",
    index=False
)

test.to_csv(
    "../data/processed/test.csv",
    index=False
)
```

# DS423 Group 7 Roadmap
## Project: E-commerce Product Recommendation

Course: DS423 – Machine Learning with Large Datasets

Dataset:
Retailrocket E-commerce Dataset
https://www.kaggle.com/datasets/retailrocket/ecommerce-dataset

Goal:
Build an e-commerce recommender system from user-item interaction events
(View, Add-to-Cart, Transaction)
and recommend Top-N products.

Evaluation Metrics

- Precision@K
- Recall@K
- NDCG
